# Task 2.2
**Using numpy.genfromtxt or numpy.loadtxt to read the above BTC-to-USD data as a numpy vector
named rates.**


```python
import numpy as np
import matplotlib.pyplot as plt
file_path = "BTC-USD-Close.csv"
rates = np.loadtxt(file_path, delimiter=',')
```

## Task 2.3
**Now for the 3rd Quarter we will display**  
* arithmetic mean ,  
* minimum,  
* the first quartile,  
* median,  
* the third quartile,  
* maximum,  
* standard deviation,  
* interquartile range.  

```python
# Extracting Q3 2023 data (
q3_rates = rates[181:273]
# Computing the required aggregates
arithmetic_mean = np.mean(q3_rates)
minimum = np.min(q3_rates)
q1 = np.percentile(q3_rates, 25)
median = np.median(q3_rates)
q3 = np.percentile(q3_rates, 75)
maximum = np.max(q3_rates)
std_dev = np.std(q3_rates)
iqr = q3 - q1

# Displaying aggregates
print("## Aggregates for Q3 2023")
print(f"Arithmetic Mean: {arithmetic_mean:.2f}")
print(f"Minimum: {minimum:.2f}")
print(f"Q1: {q1:.2f}")
print(f"Median: {median:.2f}")
print(f"Q3: {q3:.2f}")
print(f"Maximum: {maximum:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

## 2.4:
**Using Matplotlib we will draw Q2 2023 Data**

```python
# Defining days for Q3 2023 (starting from day 182 as day 1)
days = np.arange(182, 274)
# Ploting Q3 2023 BTC closing prices
plt.figure(figsize=(10, 5))
plt.plot(days, q3_rates, color='red', linestyle='-', label="BTC-USD Close Prices")
plt.title("BTC-USD Closing Prices in Q3 2023")
plt.xlabel("Days (from July 1, 2023)")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

The graph shows how Bitcoin's price changed from July to September 2023. At the start, the price stayed between $29,000 and 31,000 with small ups and downs. Around Day 220, there was a big drop below 26,000, likely due to market changes or investor reactions. After that, the price moved up and down but didnâ€™t fully recover, ending just under 27,000. This means Bitcoin had a tough time in Q3, with a sudden crash and some small recovery.

## 3.1
**Drawing a boxplot for Q3 with AM as green x**

```python
# Computing daily price changes in Q3 using numpy.diff
q3_changes = np.diff(q3_rates)

# Creating a horizontal box plot
plt.figure(figsize=(8, 4))
plt.boxplot(q3_changes, vert=False, patch_artist=True)
mean_change = np.mean(q3_changes)
plt.plot(mean_change, 1, 'gx', markersize=12, label="Mean Change")
plt.title("Distribution of BTC-to-USD Daily Price Changes in Q3 2023")
plt.xlabel("Daily Price Change (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

# Distribution of BTC-to-USD Daily Price Changes in Q3 2023

This box plot shows the **daily price changes** of Bitcoin in **Q3 2023**.

- The **blue box** represents the **middle 50% of the data** (interquartile range, IQR), meaning most daily price changes fall within this range.
- The **orange line** inside the box is the **median**, showing the middle value of the daily changes.
- The **whiskers** extend to the smallest and largest values within **1.5Ã— IQR**, covering most normal price changes.
- The **circles (outliers)** represent extreme price movements beyond the expected range, indicating sudden spikes or drops in Bitcoin's price.

### **Key Takeaways**
- Most daily price changes were within a **moderate range**, suggesting stability in BTC's price movements.
- There are **several outliers**, indicating days with **unusual price spikes or crashes**.
- The presence of negative and positive outliers suggests **high volatility** in Q3, with some days experiencing **sharp losses or gains**.

Overall, this plot highlights **Bitcoin's price fluctuations in Q3 2023**, showing both normal daily movements and exceptional price swings.


## 3.2:
**Computing no of Outliers in Q3**

```python
# Computing outliers 
q1_changes = np.percentile(q3_changes, 25)
q3_changes_75 = np.percentile(q3_changes, 75)  
iqr_changes = q3_changes_75 - q1_changes

# Defining outlier thresholds
lower_bound = q1_changes - 1.5 * iqr_changes
upper_bound = q3_changes_75 + 1.5 * iqr_changes

# Identifying outliers correctly from the actual daily changes
outliers = q3_changes[(q3_changes < lower_bound) | (q3_changes > upper_bound)]
num_outliers = len(outliers)

print(f"\nNumber of Outliers in Q3 2023: {num_outliers}")
```

### **What Do the Outliers Mean?**
Outliers in the box plot represent **unusual daily price changes** in Bitcoin during **Q3 2023**. These extreme movements could be caused by:

- **Market News & Events** â€“ Major announcements, regulations, or institutional investments.
- **Economic Factors** â€“ Interest rate changes, inflation reports, or global financial trends.
- **Market Manipulation** â€“ Large trades by big investors (whales) creating sudden spikes or drops.
- **Crypto-Specific Events** â€“ Exchange hacks, security issues, or major upgrades.

### **Key Insights**
- **More outliers** â†’ Higher volatility with frequent big price swings.
- **Fewer outliers** â†’ A more stable market with fewer unexpected moves.
- **Negative outliers** â†’ Sudden crashes due to panic selling.
- **Positive outliers** â†’ Sharp price increases from strong buying activity.

Outliers help us understand **how stable or unpredictable Bitcoin was in Q3 2023**. ðŸš€ðŸ“‰ðŸ“ˆ
